```python
def pdf_to_markdown(pdf_path, outline_as_headings=True):
  """Converts a PDF file to Markdown, using the outline for headings.

  Args:
    pdf_path: Path to the PDF file.
    outline_as_headings: Whether to use the PDF outline for headings (True) 
                          or just page numbers (False).

  Returns:
    A string containing the Markdown text.
  """

  doc = fitz.open(pdf_path)
  markdown_text = ""
  toc = doc.s
  for page in doc:
    text_box = (107, 103, 492, 744)
    page_text = page.get_textbox(text_box)
    
    # print(page_text)

    for entry in toc:
      level = "#"*entry[0]
      title = entry[1]
      if len(title) >= 6:
        page_text = page_text.replace(title, f"{level} {title}")

    markdown_text += page_text + "\n"
  
  return markdown_text

# Example usage
pdf_file = "data/raw/gruene_2021.pdf"
markdown_output = pdf_to_markdown(pdf_file)
print(markdown_output)
```
